```python
import csv
f = open("locKo.txt", 'r')
data = f.read()
a = data.replace('"','')
b = a.replace("=", ",")
c = b.split(';')
# print(c[0].lstrip())
d = []
for i in c:
  e = i.split(',')
  if len(e) == 2:
    key = e[0].lstrip().rstrip()
    if e[1] is not None:
      value = e[1].lstrip().rstrip()
    d.append([key,value])
print(d)

g = open("data2.csv", "w")
writer = csv.writer(g)
 
writer.writerows(d) ## ì—¬ê¸° ì£¼ëª©!
f.close()
  
```
